Data Collection from twitter


```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from IPython.display import display
import datetime
import os
from tqdm import tqdm

#5nd Jan 2023
start = datetime.datetime(2022, 11, 10) 
#10th Jan 2023
lastTime = datetime.datetime(2022, 11, 11)

while start != lastTime:
    finish = start + datetime.timedelta(days = 1)
    start_month = start.strftime("%m")
    start_day = start.strftime("%d")
    end_month = finish.strftime("%m")
    end_day = finish.strftime("%d")

    filename = '/content/drive/MyDrive/Thesis/Datasets/rawTweetsCollection/tweetsData-' + start_day + '-' + start_month + '-2022.csv'
    isExist = os.path.isfile(filename)

    if isExist:
        print(filename + " has already existed!")
        start = finish
        continue
        
    query = "I (did OR do OR my) lang:en until:2022-" + end_month + "-" + end_day + " since:2022-" + start_month + "-" + start_day
    print("Collected Data: " + start_day + " - " + start_month + " - 2022")
    tweets = []
    limit = 50000
    progressBar = tqdm(total = limit, position = 0, leave = True)

    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        if len(tweets) == limit:
            break
        else:
            tweets.append(tweet)
            progressBar.update(1)

    progressBar.close()  
    df = pd.DataFrame(tweets)
    display(df.shape)
    df.to_csv(filename)
    start = finish

```
